## **ต่อตาราง**

- ต่อในแนวแกน Y (เพิ่มจำนวนข้อมูล)
- ต่อในแนวแกน X (เพิ่มรายละเอียดของข้อมูล)

### **ต่อแกน Y**

```python
BKK = data_df[data_df['province_of_onset']=='กรุงเทพมหานคร']  # เลือกดูกรุงเทพ
BKK.head()    # ต้องออกมา 5 เเถว
```

```python
KKC = data_df[data_df['province_of_onset']=='ขอนแก่น']   # เลือกดูขอนเเก่น
KKC.head()   # ต้องออกมา 5 เเถว
```

### **ต่อตารางให้เป็นข้อมูล 2 จังหวัด**

```python
dataBKKKKC = pd.concat([BKK,KKC])   # ต่อตารางในเเนวเเกน Y
dataBKKKKC
```

```python
dataBKKKKC = pd.concat([BKK,KKC,BKK])     # ต้องมีจำนวนคอลัมน์เท่ากันถึงจะสามารถใช้ได้
```

```python
dataBKKKKC
```

### **ถ้าเอาตารางไม่เหมือนกันมา concat กัน**

```python
BKK2 = data_df[data_df['province_of_onset']=='กรุงเทพมหานคร'][['sex','age','nationality']]   # เลือกดู กรุงเทพ เเต่เลือกเฉพาะคอลัมน์'sex','age','nationality'
BKK2.head()
```

```python
KKC2 = data_df[data_df['province_of_onset']=='ขอนแก่น'][['age','nationality','sex']]    # เลือกดู ขอนเเก่น เเต่เลือกเฉพาะคอลัมน์ 'age','nationality','sex'
KKC2.head()
```

```python
pd.concat([BKK2,KKC2])    # ถ้าคอลัมน์เป็นชื่อเดียวกันจะ สลับที่ให้เลย
```

```python
KKC3 = data_df[data_df['province_of_onset']=='ขอนแก่น'][['age','nationality','risk']]    # เลือกดู ขอนเเก่น เเต่เลือกเฉพาะคอลัมน์ 'age','nationality','risk'
KKC3.head()
```

```python
pd.concat([BKK2,KKC3])   # มันก็จะพบค่าว่างในคอลัมน์ที่ไม่เหมือนกัน
```

```python
KKC4 = data_df[data_df['province_of_onset']=='ขอนแก่น'][['age','nationality']]    # เลือกดู ขอนเเก่น เเต่เลือกเฉพาะคอลัมน์ 'age','nationality'
KKC4.head()
```

```python
pd.concat([BKK2,KKC4])   # ตารางไม่เท่ากัน เลยมีค่าว่างในคอลัมน์ที่ตอนเเรกไม่มี
```

### **ต่อแกน X**

```python
personal_data = data_df[['No.','sex','age','nationality']]
personal_data.head()   # ตารางข้อมูลส่วนบุคคล
```

```python
covid_data = data_df[['No.','announce_date','province_of_isolation','province_of_onset','risk']].sample(frac=1)  # .sample ไว้สลับเเถวเเบบสุ่ม
covid_data.head()  # ข้อมูลโควิด
```

### **เอาตารางสองอันมาต่อกัน**

```python
personalcovid_data = personal_data.merge(covid_data)   # นำ no.1 มาเเปะ จะเอาตารางสองมาเเปะ คือ covid_data มาต่อ
personalcovid_data.head()
```

```python
personalcovid_data = covid_data.merge(personal_data)   # นำ no.1 มาเเปะ จะเอาตารางสองมาเเปะ คือ personal_data มาต่อ
personalcovid_data.head()
```

## **Save Table**

```python
personalcovid_data.to_csv(os.path.join(path,'my_processed_data.csv') )   # เซฟข้อมูลใน google drive ในโฟลเดอร์
```

## **Groupby**

https://www.kaggle.com/crawford/python-groupby-tutorial#

```python
data_df.groupby('nationality').count()   # จัดกลุ่มที่มี nationality เหมือนกัน เเล้ว ทำการนับจำนวนด้วย .count()
```

```python
data_df.groupby('nationality').mean()   # หาค่าเฉลี่ยด้วย  .mean()
```

```python
data_df.groupby('nationality')['age'].min()   #  จัดกลุ่ม nationality เเล้วดู ค่า min ของ age
```

```python
data_df.groupby('nationality').min(' ')   # ดูค่า min ของการจัดกลุ่ม nationality โดยไม่เลือกคอลัมน์ ต้องใช้ (' ')
```
