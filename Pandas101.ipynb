## Quiz ตัดตารางออกมาเฉพาะปี 2021 ```announce_date``` ในปี 2021


**Hint**
*   วนลูปหา index ของปี 2021
*   ตัดตารางมาเฉพาะ ปี 2021



````
TF=list()
for each_row in data_covid.iterrows():
    if each_row[1]['announce_date'].split('/')[2] == '2021':
        TF.append(True)
    else:
        TF.append(False)

data_covid[TF].head()
````

````
TF
````

### **HW กลุ่ม 2 || เลือกข้อมูลมาเฉพาะวัน announce_date เป็นวันที่ 1-20**
### วันที่ คือเลขตัวที่สอง

````
TF=list()   # กำหนด TF ให้เป็น list ว่าง เพื่อที่จะใช้เติมค่าใน loop
for each_row in data_covid.iterrows():
    if each_row[1]['announce_date'].split('/')[1] <= '20' :
        TF.append(True)  # ถ้าเงื่อนไขถูกต้องให้เพิ่มเข้ามาใน TF
    else:
        TF.append(False)   #ถ้าเงื่อนไขไม่ถูกต้อง จะไม่ให้เพิ่มเข้ามาใน TF

data_covid[TF]   # แสดงผล
````
